## Sequential Programming

A program executes instructions in order, from top to bottom.

The sequence of steps affects the output.

```python
print("Step 1: Start")
print("Step 2: Process")
print("Step 3: End")
```

Observation:
Code runs top to bottom. Changing order changes output.

## print() Function

Used to display output on the screen.

Rules:
- Strings must be inside quotes
- Numbers are written without quotes

```python
print("Hello, World")
print(10)
print(5 + 3)
```

```python
print("Hello", "World")
print("Hello" + "World")
print("Hi " * 3)
```

```python
print("Hello\nWorld")
print("Python", "is", "fun", sep="-")
print("Hello", end=" ")
print("World")
```

Notes:
- comma adds space
- '+' joins without space
- '*' repeats strings

## Numeric Data Types

Common types:
- int → integers (e.g., 10)
- float → decimal numbers (e.g., 10.5)

## Arithmetic Operators

Operators:
- '+' addition
- '-' subtraction
- '*' multiplication
- / division
- // floor division
- % modulus
- ** exponent

```python
a = 10
b = 3

print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a // b)
print(a % b)
print(a ** b)
```

## Errors in Python

Syntax Error:
- Occurs due to incorrect code structure

Runtime Error:
- Occurs during execution

```python
# Syntax Error example (do not run)
# print("Hello)

# Runtime Error example
print(10 / 0) #Division by zero error
```

Difference:
- syntax error = code is wrong
- runtime error = code runs but crashes
