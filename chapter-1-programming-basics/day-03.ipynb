## Loops

Loops are used to repeat a block of code multiple times.

Types:
- for loop
- while loop

## for Loop

Used when the number of iterations is known.

```python
for i in range(5):
    print(i)
```

```python
for i in range(1, 6):
    print("Number:", i)
```

range(start, end) : end - 1

## while Loop

Used when the number of iterations is not fixed.

```python
i = 1

while i <= 5:
    print(i)
    i += 1
```

Need to manually update variable, otherwise infinite loop

## break and continue

break → exits the loop  
continue → skips current iteration

```python
for i in range(5):
    if i == 3:
        break
    print(i)
```

break stops loop completely

```python
for i in range(5):
    if i == 3:
        continue
    print(i)
```

continue skips only that iteration

```python
num = int(input("Enter a number: "))

for i in range(1, 11):
    print(num, "x", i, "=", num * i)
```

```python
#even numbers
for i in range(2, 21, 2):
    print(i)
```

```python
#sum of n numbers
n = int(input("Enter n: "))
total = 0
for i in range(1, n + 1):
    total += i
print("Sum:", total)
```

```python
for i in range(1, 6):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
```

```python
for i in range(5, 0, -1):
    print("*" * i)
```

```python
n = int(input("Enter number: "))
fact = 1

for i in range(1, n + 1):
    fact *= i

print("Factorial:", fact)
```

```python
n = int(input("Enter number: "))
fact = 1
i = 1

while i <= n:
    fact *= i
    i += 1

print("Factorial:", fact)
```

```python
#Palindrome Number
num = int(input("Enter a number: "))
original = num
reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num //= 10

if original == reverse:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
#Number Guessing Game
import random

target = random.randint(1, 100)
attempts = 0

while attempts < 5:
    guess = int(input("Enter your guess: "))
    attempts += 1

    if guess < target:
        print("Higher")
    elif guess > target:
        print("Lower")
    else:
        print("Correct! Attempts:", attempts)
        break
else:
    print("Game Over! The number was:", target)
```

```python
#Snail in a Well
height = 0
day = 0

while height < 30:
    day += 1
    height += 7

    if height >= 30:
        break

    height -= 5

print("Days:", day)
```

```python
for i in range(2, 10):
    for j in range(1, 10):
        print(f"{i} x {j} = {i*j}", end="  ")
    print()
```

```python
st = "Programming"

for ch in st:
    if ch in ['a', 'e', 'i', 'o', 'u']:
        break
    print(ch)
```

```python
# Pyramid
for i in range(1, 6):
    print(" " * (5 - i) + "*" * (2*i - 1))
```

## Key Takeaways

- for loop is used for fixed iterations
- while loop is used for conditions
- break stops a loop
- continue skips an iteration
- loops help automate repetitive tasks
