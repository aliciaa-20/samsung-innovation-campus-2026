## Variables, Input, Operators, Conditionals

### Variables

Variables are used to store data.

Rules:
- Must start with a letter or underscore
- Cannot use keywords
- Case-sensitive

```python
x = 10
name = "Alicia"

print(x)
print(name)
```

## Data Types

Common data types:
- int → integer
- float → decimal
- str → string
- bool → True / False

```python
a = 10
b = 3.5
c = "Hello"
d = True

print(type(a))
print(type(b))
print(type(c))
print(type(d))
```

## input() Function

Used to take input from the user.

By default, input is taken as a string.

```python
name = input("Enter your name: ")
print("Hello ", name)
```

```python
num = input("Enter a number: ")
print(num + "5")   # string concat
```

```python
# converting input to integer
num = int(input("Enter a number: "))
print(num + 5)
```

```python
a = 10
b = 5

print(a > b)
print(a == b)
print(a != b)
```

```python
x = 10

print(x > 5 and x < 20)
print(x < 5 or x == 10)
print(not(x > 5))
```

## Conditional Statements

Used for decision making.

```python
age = int(input("Enter age: "))

if age >= 18:
    print("Eligible to vote")
else:
    print("Not eligible")
```

```python
num = int(input("Enter a number: "))

if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
''' Real-World Example

BMI Calculator logic:
- Input height and weight
- Calculate BMI
- Display result'''

height = float(input("Enter height (m): "))
weight = float(input("Enter weight (kg): "))

bmi = weight / (height ** 2)

if bmi < 18:
    print("Underweight")
elif bmi < 25:
    print("Normal")
else:
    print("Overweight")
```

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
print(a + b)
```

```python
num = int(input("Enter a number: "))
if num >= 0:
    print("Positive")
else:
    print("Negative")
```

```python
num = int(input("Enter a number: "))
if num % 5 == 0:
    print("Divisible by 5")
else:
    print("Not divisible")
```

```python
radius = float(input("Enter radius: "))
pi = 3.141592
circumference = 2 * pi * radius
area = pi * radius * radius

print("Circumference:", circumference)
print("Area:", area)
```

```python
print(int("50") + 50)   # 100
print(str(50) + "50")   # "5050"
```
