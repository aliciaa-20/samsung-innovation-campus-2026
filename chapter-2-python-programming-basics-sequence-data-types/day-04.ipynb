## Lists, Tuples, Dictionary Basics

### List

```python
# List creation
fruits = ['banana', 'apple', 'orange', 'kiwi']
print(fruits)

mixed = [100, 'apple', 200]
print(mixed)

# Indexing
print(fruits[0])
print(fruits[-1])
```

- Lists can store multiple data types  
- Indexing starts from 0  
- Negative index accesses from end  

```python
# Built-in functions
nums = [10, 20, 30, 40]

print(len(nums))
print(max(nums))
print(min(nums))
print(sum(nums))
```

```python
# append vs extend
a = [1, 2, 3]
a.append([4, 5])
print(a)

b = [1, 2, 3]
b.extend([4, 5])
print(b)
```

append → adds as a single element  
extend → adds elements individually  

```python
# delete
lst = [10, 20, 30, 40]

del lst[1] # deletes the element at index 1
lst.remove(30) # removes the first occurrence of the value 30
lst.pop() # removes the last element

print(lst)
```

```python
# slicing
a = [10, 20, 30, 40, 50, 60]

print(a[1:4])
print(a[:3])
print(a[2:])
print(a[:])
```

```python
# in operator
nums = [10, 20, 30]

print(20 in nums)
print(50 not in nums)
```

in - used for checking presence of elements

```python
# Q1 prime list
prime_list = [2, 3, 5, 7]
print(prime_list)
prime_list.append(11)
print(prime_list)
print(prime_list[0])
```

```python
# Q3 nested loop multiplication
list1 = [1, 2]
list2 = [3, 4]

for i in list1:
    for j in list2:
        print(i * j)
```

```python
numbers = [5, 10, 15, 20]

numbers.append(25)
numbers.remove(10)

print("Updated list:", numbers)
print("Max value:", max(numbers))
```

### Tuples

```python
# tuple basics
t = (1, 2, 3)
print(t)
```

tupes are immutable

```python
# unpacking
a, b = (10, 20)
print(a, b)
```

```python
# swap
a = 100
b = 200

a, b = b, a
print(a, b)
```

### Dictionary

```python
# dictionary
person = {
    "Name": "David",
    "Age": 20
}

print(person["Name"])
```

Dictionaries store data in key-value pairs

```python
# add / modify
person["Job"] = "Engineer"
person["Age"] = 21

print(person)
```

```python
# delete
del person["Age"]
print(person)
```

## What I learned

- Lists are flexible and allow modification

- append vs extend behave differently

- slicing is useful for accessing subsets

- tuples are immutable

- dictionaries store structured data using keys
