## Unit 34: Using Python Modules

### What is a Module?
A module is a Python file that contains functions, variables, or classes.

### Types:
- Standard Module → Built-in (math, random, time)
- External Module → Installed using pip

### Syntax:
import module_name

from module import function

import module as alias

### Why use modules?
- Code reuse
- Organized structure
- Easier debugging

```python
import random
import time

def roll_dice(num_dice):
    return [random.randint(1, 6) for _ in range(num_dice)]

def display_roll(player, rolls):
    print(f"{player} rolled: {rolls} → Total = {sum(rolls)}")

def play_game():
    print("Welcome to Dice Game!")
    
    num_dice = int(input("Enter number of dice: "))
    rounds = int(input("Enter number of rounds: "))
    
    user_score = 0
    comp_score = 0
    
    for i in range(1, rounds + 1):
        print(f"\nRound {i}")
        
        input("Press Enter to roll dice...")
        
        user_roll = roll_dice(num_dice)
        comp_roll = roll_dice(num_dice)
        
        display_roll("You", user_roll)
        time.sleep(1)
        display_roll("Computer", comp_roll)
        
        if sum(user_roll) > sum(comp_roll):
            print("You win this round!")
            user_score += 1
        elif sum(comp_roll) > sum(user_roll):
            print("Computer wins this round!")
            comp_score += 1
        else:
            print("It's a tie!")
    
    print("\nFINAL RESULT")
    print(f"You: {user_score} | Computer: {comp_score}")
    
    if user_score > comp_score:
        print("You are the overall winner!")
    elif comp_score > user_score:
        print("Computer wins overall!")
    else:
        print("It's a draw!")

play_game()
```
