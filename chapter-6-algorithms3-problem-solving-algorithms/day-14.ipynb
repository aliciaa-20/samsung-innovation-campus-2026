## Unit 32: Dynamic Programming

Dynamic Programming stores results of subproblems to avoid recomputation.

### Key Concepts:
- Overlapping subproblems
- Memoization
- Bottom-up approach

### Examples:
- Fibonacci
- Pascal’s Triangle

```python
def fib(n):
    dp = [0]*(n+1)
    dp[1] = 1

    for i in range(2, n+1):
        dp[i] = dp[i-1] + dp[i-2]

    return dp[n]


n = int(input("Enter n: "))
print("Fibonacci:", fib(n))
```

```python
def pascal(n):
    B = [[0]*(i+1) for i in range(n+1)]

    for i in range(n+1):
        for j in range(i+1):
            if j == 0 or j == i:
                B[i][j] = 1
            else:
                B[i][j] = B[i-1][j-1] + B[i-1][j]

    return B


n = int(input("Enter n: "))
triangle = pascal(n)

for row in triangle:
    print(row)
```

## Unit 33: Backtracking

Backtracking builds solution step-by-step and removes invalid ones.

### Key Concepts:
- Depth First Search
- Pruning
- Promising

### Example:
- N Queens Problem

```python
def is_safe(board, row, col):
    for i in range(row):
        if board[i] == col or abs(board[i] - col) == abs(i - row):
            return False
    return True


def solve(board, row, n):
    if row == n:
        print(board)
        return

    for col in range(n):
        if is_safe(board, row, col):
            board[row] = col
            solve(board, row+1, n)


n = int(input("Enter N: "))
board = [-1]*n

solve(board, 0, n)
```
