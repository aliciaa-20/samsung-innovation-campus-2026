## Spotify Data Analysis

This task uses a Spotify dataset to create a custom playlist based on features like popularity, energy, and danceability.

We use Pandas to filter and analyze the dataset.

```python
import pandas as pd

df = pd.read_csv("spotify_dataset.csv")
df.head()
```

```python
print(df.shape)      # rows, columns
df.info()            # data types
df.describe()        # stats
```

```python
df = df.dropna()   # remove missing values
```

```python
# Most popular songs
top = df.sort_values(by="popularity", ascending=False)
top[["track_name", "artists", "popularity"]].head(10)
```

```python
print(df.columns)
```

```python
# Example: High energy + danceable songs
playlist = df[(df["energy"] > 0.7) & (df["danceability"] > 0.7)]

playlist[["track_name", "artists", "energy", "danceability"]].head()
```

```python
top_songs = df.sort_values(by="popularity", ascending=False)

top_songs[["track_name", "artists", "popularity"]].head(10)
```

```python
playlist.to_csv("my_playlist.csv", index=False)
```
