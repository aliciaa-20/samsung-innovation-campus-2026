```python
s_list = ['abc', 'bcd', 'bcdefg', 'abba', 'cddc', 'opq']

shortest = s_list[0]

for s in s_list:
    if len(s) < len(shortest):
        shortest = s

print("Shortest:", shortest)
```

```python
# longest
longest = s_list[0]

for s in s_list:
    if len(s) > len(longest):
        longest = s

print("Longest:", longest)
```

```python
# multiple shortest
min_len = len(min(s_list, key=len))

for s in s_list:
    if len(s) == min_len:
        print(s)
```

```python
menu = {
    "Americano": 3000,
    "Latte": 4500,
    "Espresso": 3600
}

for key in menu:
    print(key, menu[key])

# check menu
choice = input("Enter drink: ")

if choice in menu:
    print(choice, menu[choice])
else:
    print("Not in menu")

# formatted print
for key in menu:
    print("{:15s} price : {:,}".format(key, menu[key]))
```
